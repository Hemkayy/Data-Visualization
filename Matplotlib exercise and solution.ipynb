1. Create a figure object called fig using plt.figure()

2. Use add_axes to add an axis to the figure canvas at [0,0,1,1]. Call this new axis ax.

3. Plot (x,y) on that axes and set the labels and titles to match the plot below:

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('title')
```

Create a figure object and put two axes on it, ax1 and ax2. Located at [0,0,1,1] and [0.2,0.5,.2,.2] respectively.

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,0.2,0.2])
```

Now plot (x,y) on both axes. And call your figure object to show it.

```python
ax1.plot(x,y,color='red')
ax2.plot(x,y,color='blue')
fig
```

Create the plot below by adding two axes to a figure object at [0,0,1,1] and [0.2,0.5,.4,.4]

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,0.4,0.4])
```

Now use x,y, and z arrays to recreate the plot below. Notice the xlimits and y limits on the inserted plot:

```python
ax1.plot(x,z)
ax1.set_xlabel('x')
ax1.set_ylabel('z')
ax2.plot(x,y)
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_xlim([0,1])
ax2.set_ylim([0,1])
fig
```

Use plt.subplots(nrows=1, ncols=2) to create the plot below.

lw means the thickness of the line inside the box

```python
fig,axes = plt.subplots(1,2,figsize=(14,3))
axes[0].plot(x,y, lw=4,ls='--')
axes[1].plot(x,z,color='r',lw=4)
```
